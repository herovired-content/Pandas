#### Another DataFrame to explain Boolean Indexing

```python
#---------Creating a Dictionary-----------
import random
dic = { 'Class':[random.choice(['10th Std','11th Std','12th Std']) for i in range(7)],
       'Rollno':random.sample(range(10,30),7),
       'Gender':[random.choice(['M','F']) for i in range(7)],
      'English':[random.randint(10,100) for i in range(7)],
      'Physics':[random.randint(10,100) for i in range(7)],
    'Chemistry':[random.randint(10,100) for i in range(7)],
  'Scholorship':[random.choice([True,False]) for i in range(7)]}
dic
```

```python
d = pd.DataFrame(dic)
d
```

```python
criteria_1 = d['Class']=='11th Std'
criteria_1
```

```python
d[criteria_1]
```

```python
d[(criteria_1) & d['Scholorship']]
```

```python
d
```

```python
d[(criteria_1) & (d['Gender']=='F')]
```

```python
sum(d['Gender']=='F')
```

```python
all_11th_F = d[(criteria_1) & (d['Gender']=='F')]
all_11th_F
```

```python
#----------Creating a List of Boolean Values----------
temp = [True,True,True,True,False,False,False]
temp
```

```python
d[temp]
```

```python
d['Scholorship']
```

```python
d[d['Scholorship']]
```
